To package a conda environment do the following:
- with the conda environment of interest being **deactivated**, run `conda pack -n WHICHENV -o WHICHENV.tar.gz`, (`WHICHENV` is the environment to be packaged)
- `scp WHICHENV.tar.gz USERNAME@transfer.chtc.wisc.edu:/staging/USERNAME/utils` (`USERNAME` is `pravindran` for me; the directory referred to in the command is my preference)
